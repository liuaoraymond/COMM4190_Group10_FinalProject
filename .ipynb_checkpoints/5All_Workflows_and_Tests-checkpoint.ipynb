# Workflow

```python
from typing import Annotated
from typing_extensions import TypedDict

from langgraph.graph import StateGraph, START, END
from langgraph.graph.message import add_messages
from langchain_openai import ChatOpenAI
from langgraph.checkpoint.memory import MemorySaver

from IPython.display import Image, display
import ipywidgets as widgets

from dotenv import load_dotenv
_ = load_dotenv()
```

```python
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.7)
```

```python
class State(TypedDict):
    input: str
    decision: str
    output: str
    pdf_content: str  # ADD THIS LINE
    
def llm_call_1(state: State):
    """Give me relevant job openings for a career transition"""
    prompt = f"{state.get('pdf_content', '')}\n\n{state['input']}"
    result = llm.invoke(prompt)
    return {"output": result.content}

def llm_call_2(state: State):
    """Tell me which connections to reach out to"""
    prompt = f"{state.get('pdf_content', '')}\n\n{state['input']}"
    result = llm.invoke(prompt)
    return {"output": result.content}

def llm_call_3(state: State):
    """Give me open entry level opportunities"""
    prompt = f"{state.get('pdf_content', '')}\n\n{state['input']}"
    result = llm.invoke(prompt)
    return {"output": result.content}
    
    if "transition" in input_lower:
        return {"decision": "llm_call_1"}
    elif "connections" in input_lower or "network" in input_lower:
        return {"decision": "llm_call_2"}
    elif "entry-level" in input_lower or "entry" in input_lower:
        return {"decision": "llm_call_3"}
    else:
        return {"decision": "llm_call_1"}
        
def route_decision(state: State):
    if state["decision"] == "llm_call_1":
        return "llm_call_1"
    elif state["decision"] == "llm_call_2":
        return "llm_call_2"
    elif state["decision"] == "llm_call_3":
        return "llm_call_3"
    else:
        return "llm_call_1"

router_builder = StateGraph(State)

router_builder.add_node("llm_call_1", llm_call_1)
router_builder.add_node("llm_call_2", llm_call_2)
router_builder.add_node("llm_call_3", llm_call_3)
router_builder.add_node("llm_call_router", llm_call_router)

router_builder.add_edge(START, "llm_call_router")
router_builder.add_conditional_edges(
    "llm_call_router",
    route_decision,
    {
        "llm_call_1": "llm_call_1",
        "llm_call_2": "llm_call_2",
        "llm_call_3": "llm_call_3",
    },
)
router_builder.add_edge("llm_call_1", END)
router_builder.add_edge("llm_call_2", END)
router_builder.add_edge("llm_call_3", END)

router_workflow = router_builder.compile()
```

```python
display(Image(router_workflow.get_graph().draw_mermaid_png()))
```

# Testing

```python
with open("resume.txt", "r") as file:
    pdf_content = file.read()
```

### Scenario 1: Career Transition Advisor

```python
user_question = input("What would you like to know about potential roles based on your resume? Please include the new role you're trying to transition to.")

transition_response = router_workflow.invoke({
    "input": user_question,
    "decision": "",
    "output": "",
    "pdf_content": pdf_content
})
print(transition_response["output"])
```

### Scenario 2: Social Network Analyzer

```python
user_question = input("What do you want to know about leveraging your LinkedIn connections?")
connections_info = input("Paste information about your key LinkedIn connections (names, titles, companies, or just describe them): ")

# Combine all context
combined_context = f"Resume:\n{pdf_content}\n\nLinkedIn Connections:\n{connections_info}"

entry_level_response = router_workflow.invoke({
    "input": user_question,
    "decision": "",
    "output": "",
    "pdf_content": combined_context
})
print(entry_level_response["output"])
```
